```python
from pyspark.sql import functions as F

raw_path = "/Volumes/workspace/air_quality/data"

raw_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("mode", "PERMISSIVE") \
    .option("_metadata.enabled", "true") \
    .load(f"{raw_path}/*.csv")

raw_df = raw_df.withColumn("source_file", F.col("_metadata.file_path")) \
               .withColumn("ingest_ts", F.current_timestamp()) \
               .drop("_metadata") 

print(f"Loaded {raw_df.count():,} rows")
display(raw_df.select("source_file").distinct().limit(10))
```

```python
bronze_df = (raw_df
    .withColumn("file_name", F.regexp_extract("source_file", r"([^/]+)\.csv$", 1))
    .withColumn("city_code", F.regexp_extract("file_name", r"^([A-Za-z0-9_-]+?)(?:\.\w+)?$", 1))
    .withColumn("state_code", F.upper(F.substring("city_code", 1, 2)))
)

display(bronze_df
    .select("source_file", "file_name", "city_code", "state_code")
    .distinct()
    .limit(20)
)
```

```python
col_mapping = {
    "From Date": "from_date",
    "To Date": "to_date",
    "PM2.5": "pm25",
    "PM2.5 (ug/m3)": "pm25_ugm3",
    "PM10": "pm10", 
    "PM10 (ug/m3)": "pm10_ugm3",
    "NO (ug/m3)": "no_ugm3",
    "NO2 (ug/m3)": "no2_ugm3",
    "NOx (ppb)": "nox_ppb",
    "NH3 (ug/m3)": "nh3_ugm3",
    "SO2 (ug/m3)": "so2_ugm3",
    "CO (mg/m3)": "co_mgm3",
    "Ozone (ug/m3)": "ozone_ugm3",
    "Benzene (ug/m3)": "benzene_ugm3",
    "Toluene (ug/m3)": "toluene_ugm3",
    "Eth-Benzene (ug/m3)": "ethylbenzene_ugm3",
    "MP-Xylene (ug/m3)": "mp_xylene_ugm3",
    "O Xylene (ug/m3)": "o_xylene_ugm3",
    "Xylene (ug/m3)": "xylene_ugm3",
    "Temp (degree C)": "temp_c",
    "AT (degree C)": "air_temp_c",
    "RH (%)": "relative_humidity_pct",
    "WS (m/s)": "wind_speed_ms",
    "WD (deg)": "wind_direction_deg",
    "SR (W/mt2)": "solar_radiation_wmt2",
    "BP (mmHg)": "barometric_pressure_mmhg",
    "VWS (m/s)": "vertical_wind_speed_ms"
}

exprs = []
for col in bronze_df.columns:
    new_name = col_mapping.get(col, col.replace(" ", "_").replace("(", "").replace(")", "").replace("/", "_"))
    exprs.append(f"`{col}` AS `{new_name}`")

bronze_normalized = bronze_df.selectExpr(exprs)

print("Column mapping applied:")
display(spark.createDataFrame(
    [(k, v) for k, v in col_mapping.items()], 
    ["original", "normalized"]
))
```

```python
bronze_table = "workspace.air_quality.bronze_air_quality"

(bronze_normalized.write
    .format("delta")
    .mode("overwrite")
    .option("mergeSchema", "true")
    .saveAsTable(bronze_table)
)

print(f"Bronze table created: {bronze_table}")
spark.sql(f"SELECT * FROM {bronze_table} LIMIT 5").display()
```
